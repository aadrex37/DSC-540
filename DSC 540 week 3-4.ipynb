```python
df = pd.read_csv("adult_income_data.csv")
df
##reads and loads CSV into dataframe
```

```python
names = []
with open('adult_income_names.txt','r') as f:
    for line in f:
        f.readline()
        var=line.split(":")[0]
        names.append(var)
names
##Sets up reader and creates categories
```

```python
names.append('Income')
##adds income as a category to the end of the array
```

```python
df = pd.read_csv("adult_income_data.csv",names=names)
df
##reloads dataframe with categories indexed
```

```python
df.isnull().sum()
##confirms no null or missing values
```

```python
df_sub = df[['age','education','occupation']]
df_sub
## Subset dataframe into just 3 categories
```

```python
df_sub['age'].hist(bins=20)
plt.title("Age of Adults")
plt.xlabel("Age")
plt.ylabel("Amount") 
plt.grid(True) 
##histogram of age in df
```

```python
def strip_whitespace(s):
    return s.strip()
##function to strip away whitespaces
```

```python
df_sub['education_strip']=df['education'].apply(strip_whitespace)
df_sub['education']=df_sub['education_strip']
df_sub.drop(labels=['education_strip'],axis=1,inplace=True)

df_sub['occupation_strip']=df['occupation'].apply(strip_whitespace)
df_sub['occupation']=df_sub['occupation_strip']
df_sub.drop(labels=['occupation_strip'],axis=1,inplace=True)

df_sub
##Stripped away whitespace of education and occupation in df
```

```python
df_fil=df_sub[(df_sub['age']>=30) & (df_sub['age']<=50)]
df_fil
## filtered subset to find just the people in the dataframe between the age of 30 to 50 
```

```python
total_fil=df_fil.shape[0]
total_fil
##counts how many people are in the filtered list
```

```python
df_sub.groupby(['education']).mean()
##groups by education and then find the age mean of each education level
```

```python
df_sub.groupby('occupation').describe()['age']
##groups by occupation and then find the various  statistics of each occupation
```

```python
##Oldest average for the job is Exec-managerial at an average of 41 years old
##Oldest age at the 75th percentile is ? at 61 years old, but the oldest at 75% that is defined is Priv-house-serv at 57 years old
```

```python
occupation_stats= df_sub.groupby('occupation').describe()['age']
## creates df for previous groupby
```

```python
plt.figure()
plt.barh(y=occupation_stats.index,width=occupation_stats['count'])
plt.yticks()
plt.show()
## plots using bar graph for groupby of occupations
```
