```python
import numpy as np
import pandas as pd
import sqlite3
```

```python
import statsapi##api built into library
```

```python
url='https://statsapi.mlb.com/api/{ver}/league/{leagueId}/allStarFinalVote'
```

```python
baseball=statsapi.league_leader_data('earnedRunAverage',statGroup='pitching',limit=50,season=2016)
baseball##Pull top 50 ERA leaders in baseball in 2016
```

```python
df=pd.DataFrame(baseball)
df##convert series into dataframe
```

```python
##create column titles
df.columns =['Leaderboard', 'Name', 'Team', 'ERA']
df
```

```python
df=df.set_index('Leaderboard')
df
##makes the index the ERA leaderboard placements
```

```python
teams= df.sort_values('Team')
teams
##sorts leaderboard by teams for which each player plays 
```
