```python
##merge dataframe
frames=[ale,alc,alw,nle,nlc,nlw]
df=pd.concat(frames)
df
```

```python
df=df.sort_values(by='Pct.', ascending=False)##sorts by winning percentage
df
```

```python
df_temp=df.reset_index()##reset index
df=df_temp.reset_index()##resets index again to create order
df
```

```python
df.rename(columns={"index":"Division_Standing","level_0":"Overall_Standing"}, inplace=True)
##rename newly made columns
df
```

```python
df['Overall_Standing']=df['Overall_Standing']+1
df['Division_Standing']=df['Division_Standing']+1
df
```

```python
##remove unnecessary columns
col=['Overall_Standing','Division_Standing','Teams','Pct.']
df=df[col]
df
```

```python
df=df.set_index(['Overall_Standing'])##sets index to leaderboard
df
```
