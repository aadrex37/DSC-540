```python
import pandas as pd
import numpy as np
```

```python
##import and store csv file
df=pd.read_csv('Batting1.csv')
df
```

```python
##append df with a new column, batting average
df['AVG']=df['H']/df['AB']
df
```

```python
##limit amount of entrys to specific year, 2016, and eliminates those who have less than 100 at bats 
df1= df[(df['yearID']==2016) & (df['AB']>=100)]
df1
```

```python
##removes all unnecessary columns
col=['playerID','teamID','AVG']
df2=df1[col]
df2
```

```python
lead_avg=df2.sort_values(by='AVG',ascending=False)
lead_avg
##reorganizes dataframe in order of batting average from greatest batting average to least
```

```python
temp= lead_avg.reset_index()##resets index
lead=temp.reset_index()
lead
##resets index a second time creating a new column which we will make into our leaderboard 
```

```python
lead['leaderboard'] = lead['level_0']+1 ##creates leaderboard
lead=lead.drop(['index','level_0'], axis=1) ##drop old original index and level_0 columns
lead
```

```python
##limits df to top 50 in the leaderboard
leader_avg=lead[(lead['leaderboard']<=50)]
leader_avg
```

```python
df3=leader_avg.set_index(['leaderboard'])##sets index to leaderboard
df3
```

```python
teams= df3.sort_values('teamID')
teams
```

```python
df3.to_csv("mlst2.csv")
```
